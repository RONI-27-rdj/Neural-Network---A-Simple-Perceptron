#Question 1: What is Deep Learning? Briefly describe how it evolved and how it differs from traditional machine learning

ANSWER. It is a subset of machine learning that involves the use of artificial neural networks with multiple layers to learn complex patterns in data. These neural networks are designed to mimic the structure and function of the human brain, with layers of interconnected nodes (neurons) that process and transform inputs.

**Evolution:**

1. **Early beginnings**: The concept of neural networks dates back to the 1940s, but early models were limited by computational power and data availability.
2. **Backpropagatio**n: The development of backpropagation algorithms in the 1980s enabled efficient training of neural networks.
3. **Increased computing power**: Advances in computing power and the availability of large datasets have enabled the training of deeper and more complex neural networks.
4. **Recent breakthroughs**: Recent breakthroughs in deep learning have been driven by the development of new architectures, such as convolutional neural networks (CNNs) and recurrent neural networks (RNNs), and the availability of large datasets and computational resources.

**Differences from traditional machine learning:**

1. Representation learning: Deep learning models can automatically learn representations of data, such as features or patterns, whereas traditional machine learning models often require manual feature engineering.
2. Complexity: Deep learning models can handle complex data such as images, speech, and text, whereas traditional machine learning models are often limited to simpler data types.
3. Layered architecture: Deep learning models consist of multiple layers of artificial neurons, allowing them to learn hierarchical representations of data.
4. Large datasets: Deep learning models typically require large datasets to train effectively, whereas traditional machine learning models can often perform well with smaller datasets.



# Question 2: Explain the basic architecture and functioning of a Perceptron. What are its limitations?

ANSWER. The Perceptron is a type of feedforward neural network that consists of:

1. Input layer: The input layer receives the input features or data.
2. Weights and bias: Each input is associated with a weight, and a bias term is added to the weighted sum.
3. Activation function: The weighted sum is passed through an activation function, typically a step function or a sigmoid function, to produce the output.
4. Output layer: The output layer produces the final output of the Perceptron.

**Perceptron Functioning:**

1. Weighted sum: The inputs are multiplied by their corresponding weights, and the weighted sum is calculated.
2. Bias addition: The bias term is added to the weighted sum.
3. Activation function: The result is passed through the activation function to produce the output.
4. Output: The output is compared to the target output, and the weights and bias are adjusted based on the error.

**The Perceptron has several limitations:**

1. Linear separability: The Perceptron can only learn linearly separable patterns, which means it can only separate classes that are linearly separable.
2. Binary classification: The Perceptron is typically used for binary classification problems and is not suitable for multi-class classification problems.
3. Limited capacity: The Perceptron has limited capacity to learn complex patterns or relationships in data.


# Question 3. Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.

ANSWER. Activation functions are used in neural networks to:

1. Introduce non-linearity: Activation functions introduce non-linearity into the neural network, allowing it to learn complex relationships between inputs and outputs.
2. Enable learning: Activation functions enable the neural network to learn and represent more complex patterns in data.
3. Improve model capacity: Activation functions improve the capacity of the neural network to learn and generalize from data.

Here's a comparison of Sigmoid, ReLU, and Tanh activation functions:

Sigmoid Function
- Formula: σ(x) = 1 / (1 + exp(-x))
- Range: (0, 1)
- Properties: Sigmoid function is differentiable, continuous, and monotonic.
- Use cases: Sigmoid function is often used in binary classification problems, especially in the output layer.

ReLU (Rectified Linear Unit) Function
- Formula: f(x) = max(0, x)
- Range: [0, ∞)
- Properties: ReLU function is non-differentiable at x = 0, but it is computationally efficient.
- Use cases: ReLU function is widely used in hidden layers of deep neural networks due to its simplicity and efficiency.

Tanh (Hyperbolic Tangent) Function
- Formula: tanh(x) = 2 / (1 + exp(-2x)) - 1
- Range: (-1, 1)
- Properties: Tanh function is differentiable, continuous, and monotonic.
- Use cases: Tanh function is often used in hidden layers of neural networks, especially when the data has both positive and negative values.

Comparison Points:

1. Non-linearity: All three functions introduce non-linearity into the neural network.
2. Range: Sigmoid function has a range of (0, 1), ReLU function has a range of [0, ∞), and Tanh function has a range of (-1, 1).
3. Computational efficiency: ReLU function is computationally efficient due to its simplicity.
4. Differientiability: Sigmoid and Tanh functions are differentiable, while ReLU function is non-differentiable at x = 0.



#Question 4. What is the difference between Loss function and Cost function in neural networks? Provide examples.

ANSWER. In the context of neural networks, the terms "loss function" and "cost function" are often used interchangeably, but there is a subtle difference:

1. Loss Function: A loss function measures the difference between the predicted output and the actual output for a single data point or sample. It calculates the error or loss for one example.
2. Cost Function: A cost function, on the other hand, is the average loss over the entire training dataset. It calculates the total error or loss for the entire dataset.

Examples:

1. Mean Squared Error (MSE) Loss Function: MSE calculates the average squared difference between predicted and actual values for a single data point.

    - Loss Function: MSE = (y_true - y_pred)^2
    - Cost Function: Cost = (1/n) * Σ(y_true - y_pred)^2 (averaged over the entire dataset)

2. Cross-Entropy Loss Function: Cross-entropy loss calculates the difference between predicted probabilities and actual labels for a single data point.

    - Loss Function: Loss = -Σ(y_true * log(y_pred))
    - Cost Function: Cost = (1/n) * Σ(-Σ(y_true * log(y_pred))) (averaged over the entire dataset)



#Question 5: What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.

ANSWER. Optimizers play a crucial role in neural networks by adjusting the model's parameters to minimize the loss function. Their primary goal is to find the optimal values for the model's weights and biases that result in the lowest loss.

Here's a comparison of three popular optimizers:

Gradient Descent (GD)
- Update Rule: w = w - α * ∇L(w)
- Properties: GD is a simple and intuitive optimizer that updates the parameters in the direction of the negative gradient.
- Limitations: GD can be slow and might get stuck in local minima.

Adam (Adaptive Moment Estimation)
- Update Rule: Adam adapts the learning rate for each parameter based on the magnitude of the gradient.
- Properties: Adam combines the benefits of AdaGrad and RMSprop, handling sparse gradients and non-stationary objectives.
- Advantages: Adam is widely used due to its ability to adapt to changing gradients and converge quickly.

RMSprop (Root Mean Square Propagation)
- Update Rule: RMSprop adapts the learning rate for each parameter based on the magnitude of recent gradients.
- Properties: RMSprop helps to stabilize the update process by dividing the learning rate by an exponentially decaying average of squared gradients.
- Advantages: RMSprop is effective in handling non-stationary objectives and is less sensitive to hyperparameter tuning.


#Question 6: Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.

```python
import numpy as np

# Define the inputs and outputs for the logical AND gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Initialize the weights and bias
weights = np.random.rand(2)
bias = np.random.rand(1)[0]

# Define the learning rate and number of epochs
learning_rate = 0.1
epochs = 100

# Train the perceptron
for epoch in range(epochs):
    for i in range(len(X)):
        # Forward pass
        output = np.dot(X[i], weights) + bias
        prediction = 1 if output >= 0 else 0

        # Backward pass
        error = y[i] - prediction
        weights += learning_rate * error * X[i]
        bias += learning_rate * error

# Test the perceptron
print("Logical AND Gate:")
for i in range(len(X)):
    output = np.dot(X[i], weights) + bias
    prediction = 1 if output >= 0 else 0
    print(f"{X[i][0]} AND {X[i][1]} = {prediction}")
```

#Question 7: Implement and visualize Sigmoid, ReLU, and Tanh activation functions using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Generate input values
x = np.linspace(-5, 5, 100)

# Calculate output values
y_sigmoid = sigmoid(x)
y_relu = relu(x)
y_tanh = tanh(x)
# Create a figure with subplots
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Plot Sigmoid function
axs[0].plot(x, y_sigmoid)
axs[0].set_title('Sigmoid Function')
axs[0].set_xlabel('Input')
axs[0].set_ylabel('Output')

# Plot ReLU function
axs[1].plot(x, y_relu)
axs[1].set_title('ReLU Function')
axs[1].set_xlabel('Input')
axs[1].set_ylabel('Output')

# Plot Tanh function
axs[2].plot(x, y_tanh)
axs[2].set_title('Tanh Function')
axs[2].set_xlabel('Input')
axs[2].set_ylabel('Output')

# Layout so plots do not overlap
fig.tight_layout()

# Display the plot
plt.show()
```

#Question 8: Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.

```python
# Import necessary libraries
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Load MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Preprocess the data
x_train = x_train.reshape(60000, 784).astype('float32') / 255
x_test = x_test.reshape(10000, 784).astype('float32') / 255
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the model
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, batch_size=128, epochs=5, verbose=1, validation_data=(x_test, y_test))

# Print training accuracy
print("Training Accuracy:", history.history['accuracy'][-1])


```

#Question 9: Visualize the loss and accuracy curves for a neural network model trained on the Fashion MNIST dataset. Interpret the training behavior.

```python
# Import necessary libraries
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
import matplotlib.pyplot as plt

# Load Fashion MNIST dataset
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Preprocess the data
x_train = x_train.reshape(60000, 784).astype('float32') / 255
x_test = x_test.reshape(10000, 784).astype('float32') / 255
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the model
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(512, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, batch_size=128, epochs=10, verbose=1, validation_data=(x_test, y_test))

# Plot the loss and accuracy curves
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy Curve')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

plt.tight_layout()
plt.show()
```
